#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# %matplotlib inline
%matplotlib notebook
```

```python
df = pd.read_csv('SINASC_RO_2019.csv')
df.head()
```

```python
df.info()
```

```python
var_interesse = ['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']
sinasc = df[var_interesse]
sinasc
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
fig, ax = plt.subplots(figsize=(9,4))
categorias = sinasc.columns
count = sinasc.count()

ax.bar(categorias, count)

plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
var_interesse_consultas = ['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE', 'CONSULTAS']
sinasc_consultas = df[var_interesse_consultas]
sinasc_consultas
```

```python
sinasc_consultas['APGAR5'].mean()
```

```python
sinasc_consultas['CONSULTAS'].unique()
```

```python
fig, ax = plt.subplots(figsize=(9,4))
categoria = sinasc_consultas['APGAR5'].mean()
consultas = sinasc_consultas['CONSULTAS']

ax.bar(consultas, categoria)

plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,8))
```

```python
sns.countplot(data=sinasc_consultas, x='SEXO', ax=ax[0][0])
sns.countplot(data=sinasc_consultas, x='GESTACAO', ax=ax[0][1])
sns.countplot(data=sinasc_consultas, x='ESTCIVMAE', ax=ax[1][0])
sns.countplot(data=sinasc_consultas, x='CONSULTAS', ax=ax[1][1])

ax[0][1].tick_params(axis="x", rotation = 30)
plt.subplots_adjust(hspace=0.5)
```
